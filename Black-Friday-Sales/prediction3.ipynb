# Modeling

```python
y = df['Purchase'] 
X = df.drop(['Purchase'],axis=1)
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
sc = StandardScaler()
sc.fit_transform(X_train)
```

```python
sc.fit_transform(X_test)
```

```python
lm=LinearRegression()
lm.fit(X_train, y_train)
w=lm.coef_
b=lm.intercept_
print(w)
print(b)
```

```python
predictions=lm.predict(X_test)
predictions
```

```python
RF = RandomForestRegressor(n_estimators=10,max_depth=7,n_jobs=-1,random_state=42)
RF.fit(X_train,y_train)
```

```python
y_pred = RF.predict(X_test)
y_pred
```

```python
y_test
```

```python
MAE = mean_absolute_error(y_test,y_pred)
MAE
```

```python
r2_score(y_test, y_pred)
```

```python
y.mean()
```

```python
y_test.mean()
```

```python
xgb_reg=xgboost(n_estimators=480, learning_rate=0.2, max_depth=8)
```

```python
xgb_reg.fit(X_train,y_train)
```

```python
xgb_y_pred=xgb_reg.predict(X_test)
```

```python
print("RMSE:",np.sqrt(mean_absolute_error(y_test, xgb_y_pred)))
print("y_mean",(y.mean()))
print("R2 Score:",round(r2_score(y_test,xgb_y_pred)*100,2))
```
